```python
import pandas as pd
import matplotlib.pyplot as plt
from scripts1 import project_functions
data = project_functions.load_and_process("../data/raw/AirQualityUCI.csv")
data
```

```python
data.describe()
#This describes the characteristics of the clean data, however for some reason all the min values return a "-200" value.
```

```python
import seaborn as sns

corr = data.corr()
sns.heatmap(data.corr(), annot = True, vmin=-1, vmax=1, center= 0, cmap= 'PiYG')
plt.title('Correlation Matrix between Columns',fontsize=12)
plt.show()

#Correlation Matrix between all the columns
#1. there is a good amount of correlation >.7 in Tin Oxide Column between other columns
#2. Relative Humidity also has a lot of high correlations between columns, along with Temperature and Absolute Humidity
```

```python
sns.pairplot(data)

#This set of pairplots and the previous correlation matrix are the same essentially
#As correlation increases, the concentration of the plotted points on the scatterplot increase and are more densely "packed"
#And vice versa, this relationship holds true as correlation decreases as shown in the the "CO(GT)" and "Absolute Humidity" pairplot
#We see that the plots are extremely far apart although theyre still densely "packed" we can see the correlation is negative
#Due to the far distance between the 4 clusters of the plot.
```
