```python
import seaborn as sns



fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(data.corr(), ax=ax, annot = True, vmin=-1, vmax=1, center= 0, cmap= 'PiYG')
plt.title('Correlation Matrix between Columns',fontsize=15)
plt.show()

#Correlation Matrix between all the columns
#1. There is a strong positive correlation between almost all the readings of the gasses with a majority being >0.5
#2. However there is strong negative correlation between the Tungsten Oxide (nominally NOx Targeted in Sensor 3)...
#2 conti'd. between all the other gas readings. >-0.5
#3. Weak correlation between the Temperature and Absolute Humidity, but a strong correlation would show...
#3 cont'd. if a larger temperature range was given to show its relation (higher temp. = higher abs. humidity)
#4. Tungsten Oxide (NO2,S4) showed a .52 correlation with Absolute Humidity, unsure why, but many speculations can be made
```

```python
sns.pairplot(data)

#This set of pairplots is essentially the above Correlation Matrix plotted for better visualization of correlation relationships
#As correlation increases, the concentration of the plotted points on the scatterplot increase and are more densely "packed"
#And vice versa.
#
#Positive correlations show a positive linear relationship as expected as shown in the majority of gas concentraion pair plots
#And negative correlations show a negative relationship as shown in the Tungsten Oxide (NOx, S3) pairs
#Weak correlations can be shown in the Temperature, Relative Humidity, Absolue Humidity pairplots by their scattered
#"cloudy" plots
#
#An interesting thing to note is the negative logarithmic curve for the Tungsten Oxide (NOx, S3) pairs, and becomes more 
#defined as the correlation values comes closer to "-1"
```
