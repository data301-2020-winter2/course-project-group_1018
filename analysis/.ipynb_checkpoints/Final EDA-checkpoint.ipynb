```python
from scripts1 import project_functions
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

data = project_functions.load_and_process("../data/raw/AirQualityUCI.csv")
data #loading the dataset to make sure it works
```

```python
data.describe()
#This describes the characteristics of the clean data
#CO(GT):                       True hourly averaged concentration CO in mg/m^3 (reference analyzer)
#Tin Oxide (CO, S1):           PT08.S1 (tin oxide) hourly averaged sensor response (nominally CO targeted), Sensor 1
#Non Metal HydroCarbon (GT):   averaged overall Non Metanic HydroCarbons concentration in microg/m^3 (reference analyzer)
#Benzene(GT):                  True hourly averaged Benzene concentration in microg/m^3 (reference analyzer)
#Titania (NMHC, S2):           PT08.S2 (titania) hourly averaged sensor response (nominally NMHC targeted), Sensor 2
#NOx (GT):                     True hourly averaged NOx concentration in ppb (reference analyzer)
#Tungsten Oxide(NOx, S3):      PT08.S3 (tungsten oxide) hourly averaged sensor response (nominally NOx targeted), Sensor 3
#NO2 (GT):                     True hourly averaged NO2 concentration in microg/m^3 (reference analyzer)
#Tungsten Oxide(NO2, S4):      PT08.S4 (tungsten oxide) hourly averaged sensor response (nominally NO2 targeted), Sensor 4
#Indium Oxide (O3, S5):        PT08.S5 (indium oxide) hourly averaged sensor response (nominally O3 targeted), Sensor 5

#Ground Truth (GT),  hourly averaged concentrations
```

```python
fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(data.corr(), ax=ax, annot = True, vmin=-1, vmax=1, center= 0, cmap= 'PiYG')
plt.title('Correlation Matrix between Columns',fontsize=15)
plt.show()

#Correlation Matrix between all the columns
#1. There is a strong positive correlation between almost all the readings of the gasses with a majority being >0.5
#2. However there is strong negative correlation between the Tungsten Oxide (nominally NOx Targeted in Sensor 3)...
#2 conti'd. between all the other gas readings. >-0.5
#3. Weak correlation between the Temperature and Absolute Humidity, but a strong correlation would show...
#3 cont'd. if a larger temperature range was given to show its relation (higher temp. = higher abs. humidity)
#4. Tungsten Oxide (NO2,S4) showed a .52 correlation with Absolute Humidity, unsure why, but many speculations can be made
```

```python
sns.pairplot(data)

#This set of pairplots is essentially the above Correlation Matrix plotted for better visualization of correlation relationships
#As correlation increases, the concentration of the plotted points on the scatterplot increase and are more densely "packed"
#And vice versa.
#
#Positive correlations show a positive linear relationship as expected as shown in the majority of gas concentraion pair plots
#And negative correlations show a negative relationship as shown in the Tungsten Oxide (NOx, S3) pairs
#Weak correlations can be shown in the Temperature, Relative Humidity, Absolue Humidity pairplots by their scattered
#"cloudy" plots
#
#An interesting thing to note is the negative logarithmic curve for the Tungsten Oxide (NOx, S3) pairs, and becomes more 
#defined as the correlation values comes closer to "-1"
```

```python
#convert time of day to int and sum up yearly concentration into one value

data2 = pd.DataFrame(columns = ["CO(GT)","NOx(GT)", "Benzene(GT)", "Non Metal HydroCarbon (GT)", "Hour of Day"])
data3 = pd.DataFrame(columns = ["CO(GT)","NOx(GT)", "Benzene(GT)", "Non Metal HydroCarbon (GT)", "Hour of Day"])

data3["Hour of Day"] = np.arange(start=0, stop=24, step=1)
data2['Hour of Day'] = data['Hour of Day'].apply(lambda s: int(s.split('.')[0]))
data2["Benzene(GT)"], data2["NOx(GT)"], data2["Non Metal HydroCarbon (GT)"], data2["CO(GT)"] = data["Benzene(GT)"], data["NOx(GT)"], data["Non Metal HydroCarbon (GT)"], data["CO(GT)"]

#put the sums in one place
s = 0
for i in range(24):
    s=data2.loc[data2['Hour of Day'] == i]["Benzene(GT)"].sum()
    data3.loc[i, "Benzene(GT)"]= s
    s=data2.loc[data2['Hour of Day'] == i]["NOx(GT)"].sum()
    data3.loc[i,"NOx(GT)"]= s
    s=data2.loc[data2['Hour of Day'] == i]["Non Metal HydroCarbon (GT)"].sum()
    data3.loc[i, "Non Metal HydroCarbon (GT)"]= s
    s=data2.loc[data2['Hour of Day'] == i]["CO(GT)"].sum()
    data3.loc[i, "CO(GT)"]= s

#set up subplots
fig, axes = plt.subplots(nrows=2, ncols=2)
fig.set_figheight(10)
fig.set_figwidth(15)

#graph data
df = data3.filter(['NOx(GT)'], axis=1)
df.plot(ax=axes[0,0])
df = data3.filter(['Benzene(GT)'], axis=1)
df.plot(ax=axes[0,1])
df = data3.filter(['Non Metal HydroCarbon (GT)'], axis=1)
df.plot(ax=axes[1,0])
df = data3.filter(['CO(GT)'], axis=1)
df.plot(ax=axes[1,1])
```
